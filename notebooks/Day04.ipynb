# Day 4: 辞書とセット

## Learning Objectives
- 辞書の基本操作を習得する
- セットの特性を理解する
- 画像処理への応用を学ぶ

```python
# 辞書の基本
person = {
    "name": "Alice",
    "age": 20,
    "city": "Tokyo"
}

print(person["name"])
print(person.get("age"))
print("name" in person)
```

```python
# 辞書の操作
person["email"] = "alice@example.com"
person["age"] = 21
print(person)

for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# セットの基本
colors = {"red", "green", "blue"}
numbers = {1, 2, 3, 2, 1}
print(numbers)  # {1, 2, 3}

# 集合演算
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(f"和集合: {a | b}")
print(f"積集合: {a & b}")
print(f"差集合: {a - b}")
```

```python
# 画像処理への応用：色のカウント
pixels = [(255, 0, 0), (0, 255, 0), (255, 0, 0), (0, 0, 255)]
color_count = {}

for pixel in pixels:
    color_count[pixel] = color_count.get(pixel, 0) + 1

for color, count in color_count.items():
    print(f"{color}: {count}回")
```

## Exercise: 画像の色パレット分析

```python
pixels = [
    (255, 0, 0), (0, 255, 0), (0, 0, 255),
    (255, 0, 0), (255, 0, 0), (0, 255, 0),
    (128, 128, 128), (255, 255, 255), (0, 0, 0)
]

# ヒストグラムを作成
histogram = {}
for pixel in pixels:
    histogram[pixel] = histogram.get(pixel, 0) + 1

for color, count in sorted(histogram.items()):
    print(f"{color}: {count}回")
```
